#### L'apprentissage supervise

L'apprentissage supervise se fait sur des donnees "labelisees", autrement dit, la caracteristique (e.q. "feature") qui nous interesse est clairement marquee sur les donnees que nous avons. Le but etant de ce baser sur ces donnees etiquettees pour etiquetter des donnees qui ne le sont pas.

On dit que cet apprentissage est supervise car on associe des "bonnes reponses" concretes aux elements de notre dataset.

<img src="../assets/img/intro/label_unlabel.png"/>

<img src="../assets/img/intro/supervised_detailed.png"/>

#### L'apprentissage non-supervise

L'apprentissage non-supervise se fait sur des donnees non "labelisees". Le but de l'algorithme est de trouver de l'information pertinnente sans pouvoir s'appuyer sur un jeu de "bonnes reponses". En regle generale les algorithmes non-supervise servent a regrouper ou filtrer des elements en fonction de leur similaritees ou differences.

<img src="../assets/img/intro/unsupervised.png"/>

Un example concret serais un auto-tri d'emails. On pourrais trier les emails en fonction de leur poids et du nombre de destinataires et remarquer que ca forme des clusters (e.q. groupes). En fonction de vos emails vous pourrez vous retrouver un petit ou un grand nombres de clusters differents ayant une significati

<img src="../assets/img/intro/unsupervised_use.png"/>

##### Clustering n'est pas classification

Notez que le clustering rassembles des elements similaires, tandis que la classification rassemble des elements dans les groupes que vous avez vous meme definis. C'est pour ca que dans l'example des emails ci-dessus n'ont pas de signification forte, ils sont similaires mais un email du cluster 3 ne garantis pas qu'il s'agisse d'un email promotionnel, d'un spam, d'une lettre de depart ou d'un memo. Pour cela il faudrait utiliser un algo de classification entrainne avec des example pour chaque type d'email.

#### L'apprentissage par renforcement

Le reinforcement learning ne se base pas sur des datasets, on va plutot definir ce que l'ordinateur peu faire ainsi que la liste des resultat que l'on veux et la liste des resultats que l'on ne veux pas. Une fais que c'est fait on programme l'ordinateur pour tester plein de combinaisons dans les limites de ce qu'il a le droit de faire et garder ce qui donne des resultat positif et jeter ce qui donne des resultats negatif.

### les deux phases de vie d'un modele (apprentissage puis utilisation)

> 1.A Decision Process: In general, machine learning algorithms are used to make a prediction or classification. Based on some input data, which can be labelled or unlabeled, your algorithm will produce an estimate about a pattern in the data.
>
> 2.An Error Function: An error function serves to evaluate the prediction of the model. If there are known examples, an error function can make a comparison to assess the accuracy of the model.
>
> 3. An Model Optimization Process: If the model can fit better to the data points in the training set, then weights are adjusted to reduce the discrepancy between the known example and the model estimate. The algorithm will repeat this evaluate and optimize process, updating weights autonomously until a threshold of accuracy has been met.
