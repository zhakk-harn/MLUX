#### L'apprentissage supervise

L'apprentissage supervise se fait sur des donnees "labelisees", autrement dit, la caracteristique (e.q. "feature") qui nous interesse est clairement marquee sur les donnees que nous avons. Le but etant de ce baser sur ces donnees etiquettees pour etiquetter des donnees comparables qui ne le sont pas.

On dit que cet apprentissage est supervise car on fourni des "bonnes reponses" concretes aux elements de notre dataset.

<img src="../assets/img/intro/label_unlabel.png"/>

<img src="../assets/img/intro/supervised_detailed.png"/>

#### L'apprentissage non-supervise

L'apprentissage non-supervise se fait sur des donnees non "labelisees". Le but de l'algorithme est de trouver de l'information pertinnente sans pouvoir s'appuyer sur un jeu de "bonnes reponses". En regle generale les algorithmes non-supervise servent a regrouper ou filtrer des elements en fonction de leur similaritees ou differences.

<img src="../assets/img/intro/unsupervised.png"/>

Un example concret serais un auto-tri d'emails. On pourrais trier les emails en fonction de leur poids et du nombre de destinataires et remarquer que ca forme des clusters (e.q. groupes). En fonction de vos emails vous pourrez retrouver un petit ou un grand nombres de clusters differents.

<img src="../assets/img/intro/unsupervised_use.png"/>

##### Clustering n'est pas classification

Notez que le clustering rassembles des elements similaires, tandis que la classification rassemble des elements dans les groupes que vous avez vous meme definis. C'est pour ca que dans l'example ci-dessus les emails n'ont pas de signification particuliere, ils sont similaires mais un email du cluster 3 ne garantis pas qu'il s'agisse d'un email promotionnel, d'un spam, d'une lettre de depart ou d'un memo. Pour cela il faudrait utiliser un algo de classification entrainne avec des example pour chaque type d'email.

#### L'apprentissage par renforcement

Le reinforcement learning ne se base pas sur des datasets, on va plutot definir ce que l'ordinateur peu faire ainsi que la liste des resultat que l'on veux et la liste des resultats que l'on ne veux pas. Une fois que c'est fait on programme l'ordinateur pour tester plein de combinaisons dans les limites de ce qu'il a le droit de faire et garder ce qui donne des resultat positif et jeter ce qui donne des resultats negatif.

Imaginons une simulation dans laquelle nous avons notre modele representer par le robot en bas a gauche. Dans notre simulation le robot peux se deplacer de case en case, il ne peux pas aller sur une case montagne, la case tresor lui fait gagner des points et la case dragon va lui faire perdre la partie.

<img src="../assets/img/intro/reinforcement_learning.png"/>

Au debut, on va essayer beaucoup de combinaisons de haut, bas, gauche, ou droite jusqu'a soit tomber sur le tresor, dans ce cas la on va enregistrer la combinaison qui nous a permis d'y arriver, soit de tomber sur le dragon, auquel cas on enregistrera le chemin pour ne jamais le refaire.

Ceci dit, il nous manque une petite chose. Si le seul moyen de gagner des points est de trouver le tresor alors tous les chemins se valent. Ce que l'on peut faire est de rendre le gain du tresor dependant du nombre de case parcourue pour aller le chercher, du coup, une solution qui le trouve en 8 coups sera retenue comparee a une solution qui le trouve en 12.

(en pratique on fait un peu differement parce que ca devient vite une solution innefficace, la ca va parce que notre monde fait 4 x 4 cases.)
