## Regressions lineaire

La regression lineaire est une methode efficace et tres repandue pour estimer des valeur comme la valeur d'une maison, la valeur d'une action en bourse, l'esperance de vie d'un individu, ou le temps qu'un utilisateur va passer a regarder une video sur votre site web. On peut voir la regression lineaire sous une forme plus graphique que mathematique. C'est ce que l'on va faire en visualisant des points et une ligne qui bougent.

Ces points forment vaguement une ligne.

Le but de la regression lineaire est de tracer une droite qui passe au plus proche de tous les points. Un peu comme ca :

Pour ca on va voir comment utiliser la regression lineaire pour predire le prix de maisons via un dataset.

Si on veux vendre une maison, le mieux serais de se baser sur les autres maisons qui sont sur le marche. On peux regarder leurs features qui pourrais influencer le prix de la maison comme sa taille, le nombre de pieces, son emplacement, les ecoles et les commerces a proximite. Au final on voudrais juste une formule qui nous rend le prix que vaux notre maison ou au moins une estimation plutot correct.

On va faire ca avec un example tres simple. On va seulement se baser sur le nombre de pieces. Notre maison en a 4 et il y a 6 maisons a vendre dans le coin qui ont une, deux, trois, cinq, six et sept pieces.

D'instinct vous allez dire que la maison avec 4 pieces vaux 300. Et vous auriez raison. On peut voir une logique et faire une deduction basee dessus. Ce que vous venez de faire de manier intuitive est une regression lineaire. On peux remarquer qu'a chaque fois qu'on rajoute une piece on rajoute 50 au prix, en fait on peut voir le prix de base de la maison comme etant de 100 puis de 50 par piece, ce qu'on pourrais ecrire comme ca 

Prix = 100 + 50*(nombre de pieces)

Ce qu'on viens de faire avec cette ecriture est de specifier un *modele* qui donne une *prediciton* basee sur une *feature*, ici le nombre de pieces. Si on refait une tour rapide des definitions ca nous donne
> **feature**  
Les features d'une point de donnee sont les proprietes sur lequelles on va utiliser pour faire nos predictions. Ici on en a qu'une seule, le nombre de pieces dans la maison

> **label**  
La feature que l'on essaye de predire. Dans ce cas c'est le prix de la maison.

> **model**  
La regle, ou la formule, qui predit un label grace aux features. Dans ce cas, le modele est la formule qui nous permet de calculer le prix de la maison.

> **prediction**  
Le resultat donnee par le modele, ici le modele donne 300 comme prix si le nombre de pieces est de 4 donc "300" est la prediction

### Formuler une regle qui estime le prix des maisons
La formule utilisee pour le dataset precedent devrait encore etre correct quoique un peu faux donc on va l'ecrice comme ca

Prix = 100 + 50*(nombre de pieces) + (marge d'erreur)

On peux voir qu'a chaque fois qu'on se deplace d'un x vers la droite la ligne formee par les maisons monte de 50 et que cette ligne traverse l'axe y a la valeur 100. On dit que 50 est le *coefficient directeur* et que 100 est l'*ordonee a l'origine*. Pour faire plus court j'utiliserais "pente" et "OO"

> Coefficient directeur / pente  
Le coefficient directeur dicte l'angle que prend la ligne. On peut le retrouver en regardant de combien la ligne monte ou descent lorseque l'on se deplace d'une unitee vers la droite. Si la ligne descent le coefficient sera negatif, si elle monte il sera positif.Si la ligne est horizontale le coefficient est de 0 . Ca nous dira a quel point la valeur du label va changer si on change la valeur de la feature.

> Ordonee a l'origine / OO  
L'ordonee a l'origine est la valeur a laquelle la droite intersect avec l'axe vertical (les ordonees). Il s'agit de la valeur du label lorseque la feature est absente.

> Equation lineaire  
Il s'agit de l'equation representant une ligne, elle a 2 composantes, un coefficient directeur et une ordonee a l'origine. Pour une pente *a* et une ordonee a l'origine *b* on peux representer la ligne comme ceci y = ax + b. En machine learning, *x* est la valeur de la feature et *y* est la prediciton pour le label.

### Qu'est ce qu'il se passe si on veux utiliser plus qu'une feature? La regression multi-lineaire
Si on imagine qu'on veux predire le prix d'une maison en se basant pas seulement sur le nombre de piece mais aussi sa taille, la proximite avec des ecoles et l'age de la maison notre model de regression lineaire reste utile. Pour ca il faut utiliser chaque feature avec sa propre pente ce qui nous donnerais un model qui pourrais ressembler a ca  

Prix = 30(nombre de pieces) + 1.5(taille) + 10(proximite des ecoles) - 2(age) + 50

Tout est positif sauf le coefficient devant l'age de la maison? Parce que plus l'age de la maison est important plus son prix baisse, le coefficient est donc negatif. On dit que la correlation entre le prix et l'age est negative tandis que pour les autres parametre elle est positive.  
Qu'est ce qu'il se passe si le coefficient est de 0 plutot que 10, 1.5 ou 30? Ca voudrait dire que la feature en question n'as pas d'influence sur le prix. Au contraire plus le coefficient est un grand nombre et plus la feature est influente.

Pendant ce chapitre on va visualiser notre modele comme une droite mais pendant la visualisation chaque colonne de notre dataset represente une dimmension donc si on a un dataset avec deux colonnes on peux le representer sur un graph avec 2 axes. Un dataset avec 3 colonnes sera represente comme une collection de points dans un graph a 3 axes, donc en 3 dimensions, a ce stade le modele sera represente comme un plan qui doit se rapprocher au maximum de tous les points. Pour ce qui est de retrouver la ligne/modele via l'ordinateur on va s'y prendre pas a pas, comme tres souvent en machine learning. On va commencer par utiliser une ligne completement aleatoire (ou avec une pente et une OO nulles) et on va la rapprocher des points petit a petit jusqu'a avoir une ligne satisfaisante.

La version simpliste est de prendre un point au hasard et de rapprocher la ligne, puis refaire la meme avec un autre points et ainsi de suite. Ce qui nous donnerais la procedure suivante :  

Entrees : un dataset de points
Sortie : une ligne qui passe proche des points
Procedure :
- prendre une ligne au hasard
- repeter :
    - prendre un point du dataset au hasard 
    - bouger la ligne pour la rapprocher du point
- resultat : la ligne

Si on repete suffisamment de fait le rapprochement de la ligne on se retrouvera avec une ligne plutot proche de tous les points.

Pour rentrer un peu plus dans le detail on va recommencer a utiliser des formules. Pour faire plus simple on va definir les raccourcis suivant
- y: le prix d'une maison dans le dataset
- y': la predicition du prix pour une maison
- x : le nombre de pieces
- a : le prix d'une piece
- b : la prix de base d'une maison avant de compter les pieces

On retrouve bien l'equation de notre droite

y' = ax + b

Ce qui revient toujours a 

Prediction du prix = (prix par piece)(nombre de pieces) + prix de bases avant les pieces

Pour commencer a se faire une idee de la procedure de la regression lineaire, prennons un modele dans lequel les pieces valent 40 et le prix de base d'une maison est de 50 nous donnans y' = 40x + 50  
Imaginons maintenant que notre dataset contienne une maison avec 2 pieces qui coute 150. Ce modele predire qu'elle vaut 40\*2+50=130. C'est pas loin mais c'est pas ca. Vu que l'estimation tombe en dessous on pourrais ameliorer le modele en augmentant un peu le prix des pieces et le prix de base d'une maison. Si on augment le prix par piece de 0.50 et la prix de base de 1 ca nous donne un nouveau modele qui predit 40.5\*2+51=132.  
C'est plus proche de 150, l'erreur est plus faible et donc le modele est meilleur ... pour ce point de donnee seulement. 

Si on veux pouvoir rapprocher une ligne il va falloir comprendre comment la manipuler. On disait que le coefficient represente la montee ou la descente de la ligne quand augment sur l'axe des x (on va a droite). Si mon coefficient directeur est de 0.5, je m'attend a ce que pour chaque x ma ligne monte de 0.5y et si mon coefficient directeur est de 2 je m'attend a ce que pour chaque x ma ligne monte de 2y. Autrement dit, augmenter le coefficient directeur fait tourner la droite dans le sens contraire des aiguille d'une montre et a l'inverse le reduire fait tourner la droite dans le sens des aiguilles d'une montre.

Si l'ordonnee a l'origine est l'intersection entre la ligne et l'axe y alors l'augmenter revient a decaler la droite vers le haut et le reduire reviens a decaler la ligne vers le bas.

Autrement dit, l'OO nous donne la position de la ligne et le coefficient directeur nous donne sa direction.

### La verison naive
On disait que notre algorithme de regression lineaire consistait a continuellement rapprocher la ligne d'un point puis d'un autre et ainsi de suite. On va donc voir comment faire tourner et glisser la droite plus pres dans tous les differents cas.

<img src="../assets/img/regression_lineaire/closer.png"/>

Pour ca il va d'abord falloir trouver ou est le point par rapport a la droite. Si le point est sous la ligne il va falloir glisser vers le bas, si il est au dessus il va falloir glisser vers le haut. Pour tourner ca va etre un peu plus complique parce que la ligne tourne sur son OO. On va avoir besoin de savoir si le point est au dessus a gauche ou en dessous a droite de l'OO, ce qui nous indiquerais qu'il faut tourner dans le sens des aiguilles d'une montre. A l'inverse si le point est au dessus a droite ou en dessous a gauche il va falloir tourner dans le sens contraire. Donc :  

- Cas 1 : Le point est au dessus de la ligne et a gauche de l'axe y, on tourne dans le sens antihoraire et on glisse vers le haut
- Cas 2 : Si le point est au dessus de la ligne et a droite de l'axe y, on tourne dans le sens horaire et on glisse vers le haut
- Cas 3 : Si le point est en dessous de la ligne et a droite de l'axe y, on tourne dans le sens horaire et on glisse vers le bas
- Cas 4 : Si le point est en dessous de la ligne et a gauche de l'axe y, on tourne dans le sens antihoraire et on glisse vers le bas

<img src="../assets/img/regression_lineaire/rotate_translate.png"/>

Maintenant que ca c'est clair, on va pouvoir detailler notre algorithme

pour rappel notre droite ressemble a : (y' = ax+b) ou :
- y' est la prediction du prix d'une maison
- a est le coefficient directeur de la droite et le prix d'une piece
- x est le nombre de piece pour lequel on souhaite avoir une prediciton
- b est l'ordonnee a l'origine soit le prix de la maison avant de prendre en compte le nombre de pieces
- (x,y) est un point du dataset qui fait correspondre un prix y a un nombre x de pieces

Entrees : 
- une ligne ayant une pente a, et une OO b, d'equation y' = ax+b
- un point de coordone (x,y)
Sortie : 
- une ligne d'equation y' = a'x+b' ou a' et b' sont un nouveau coefficient directeur et OO qui rapproche la droite du point
Procedure :
- prendre deux petit nombres n1 et n2
- si : le point est au dessus de la ligne et a droite de l'axe y
    - ajouter n1 a la pente tel que a' = a + n1
    - ajourte n2 a l'OO tel que b' = b + n2
- si : le point est au dessus de la ligne et a gauche de l'axe y
    - retirer n1 a la pente tel que a' = a - n1
    - ajouter n2 a l'OO tel que b' = b + n2
- si : le point est en dessous de la ligne et a droite de l'axe y
    - retirer n1 a la pente tel que a' = a - n1
    - retirer n2 a l'OO tel que b' = b - n2
- si : le point est en dessous de la ligne et a gauche de l'axe y
    - ajouter n1 a la pente tel que a' = a + n1
    - retirer n2 a l'OO tel que b' = b - n2
- resultat : y' = a'x + b'

Dans notre example modifier *a* reviens a modifier le prix d'une piece te modifier *b* reviens a modifier le prix de base d'une maison. Vu que le nombre de pieces ne peux pas etre negatif on peux simplifier la procedure en oubliant les cas 1 et 3.

En principe cette methode fonctionne et elle a le merite d'etre simple mais ce n'est pas ce qui est utilise car elle n'est vraiment pas efficace. En plus de ca :
- comment est ce que l'on choisi des bonnes valeur pour n1 et n2?
- est ce qu'on pourrais pas remplacer les 4 cas par un meme etape qui marche tous le temps?
On peut! Et pour ca on va voir les methode qu'on appellera "square" pour le moment.

### la version "square"
Au final le but de nos 4 cas etait de trouver le bon signe pour (+ ou -) pour changer les parametres *a* et *b* de la droite pour se rapprocher du point.

Commencons avec le deplacement. Si le point (x,y) est au dessus de la droite, alors y - y' sera positif parce que y > y', en revanche si le point est en dessous alors y - y' sera negatif parce que y < y'.

<img src="../assets/img/regression_lineaire/prediction_delta.png"/>

Donc on peut juste ajouter (y - y') a l'OO *b* pour faire glisser la droite dans la bonne direction. Le probleme est que (y - y') est possiblement trop grand et resultera en un glissement trop violent et donc a ne jamais stabiliser une droite correcte. C'est pour ca qu'on va devoir parler de "rythme d'apprentissage"

> Rythme d'apprentissage / Learning Rate  
Un petit nombre qu'on selectionne avant d'entrainer notre modele. Ce nombre nous permet de reduire la taille des changements a chaque etape. On le notera *n*

Parce que *n* est petit on peux maintenant ajouter n(y-y') a l'OO *b* sans probleme.

Pour ce qui est de modifier la pente c'est tres similaire.  
Un point (x,y) est a gauche de l'axe y si *x* < 0 et a droite si *x* > 0

<img src="../assets/img/regression_lineaire/x_to_y.png"/>

donc x(y-y') est positif si *x* et (y-y') sont tous les deux positifs ou tous les deux negatifs comme *a'* dans les cas 1 et 4. Tandis que si seulement *x* ou seulement (y-y') est positif alors x(y-y') sera negatif comme *a'* dans le cas 2 et 3. Du coup si on ajoute x(y-y') aux coefficient directeur *a* alors la droite tournera dans le bon sens. On oublis pas de le multiplier par le learning rate pour avancer a pas mesures et on est bon. nx(y-y') nous fera toujours tourner la droite dans le bon sens.

Ce qui nous donne la nouvelle procedure 
Entrees : 
- une ligne ayant une pente a, et une OO b, d'equation y' = ax+b
- un point de coordone (x,y)
- un petit nombre positif *n* (le learning rate)
Sortie : 
- une ligne d'equation y' = a'x+b' ou a' et b' sont un nouveau coefficient directeur et OO qui rapproche la droite du point
Procedure :
- ajouter nx(y-y') au coefficient directeur pour mettre a jour la direction de la droite, a' = a + nx(y-y')
- ajouter n(y-y') a l'OO pour deplacer la droite, b' = b + n(y-y')
- resultat : y' = a'x + b'

Maintenant on va pouvoir coder ca pour l'utiliser

```python
def square_technique(base_price, price_per_room, num_rooms, price, learning_rate):
    predicted_price = base_price + price_per_room*num_rooms
    base_price += learning_rate*(price-predicted_price)
    price_per_room += learning_rate*num_rooms*(price-predicted_price)
    return price_per_room, base_price
```

### alternative : la version absolue
Il existe une autre methode qu'on appellera "absolue" qui est tout aussi utilisable et un peu plus simple que la square.  
Dans la square on utilise y-y' (le difference entre la predicition et la vraie valeur) et x (le nombre de pieces) pour couvrir tous les cas en un. Mais on peut se baser uniquement sur x pour passer de 4 cas a 2. Voici la procedure

Entrees : 
- une ligne ayant une pente a, et une OO b, d'equation y' = ax+b
- un point de coordone (x,y)
- un petit nombre positif *n* (le learning rate)
Sortie : 
- une ligne d'equation y' = a'x+b' ou a' et b' sont un nouveau coefficient directeur et OO qui rapproche la droite du point
Procedure :
- si : le point est au dessus de la ligne
    - ajouter nx au coefficient directeur, a' = a + nx (on tourne dans le sens antihoraire si le point est a droite de l'axe y ou dans le sens horaire si x est a gauche de l'axe y)
    - ajouter n a l'OO tel que b' = b + n
- si : le point est en dessous de la ligne
    - retirer nx a la pente tel que a' = a - nx (on tourne dans le sens horaire si le point est a droite de l'axe y ou dans le sens antihoraire si est a droite de l'axe y) 
    - retirer n a l'OO tel que b' = b - n
- resultat : y' = a'x + b'

et voici le code qui va avec

```python
def absolute_technique(base_price, price_per_room, num_rooms, price, learning_rate):
   predicted_price = base_price + price_per_room*num_rooms
   if price > predicted_price:
       price_per_room += learning_rate*num_rooms
       base_price += learning_rate
   else:
       price_per_room -= learning_rate*num_rooms
       base_price -= learning_rate
   return price_per_room, base_price
```

### L'algorithme de regression lineaire : repeter le square ou absolute
Le plus dur est fait. Il nous reste plus qu'a prendre une droite au hasard et tous les points pour les faire passer en boucle dans l'une des deux technique qu'on viens de voir jusqu'a ce que la ligne soit bien ajustee.  

Voila la procedure

Entrees : 
- un dataset de maisons ayant chaqune un prix et un nombre de pieces
Sortie : 
- un modele pouvant predire le prix en fonction du nombre de pieces
Procedure :
- prendre une droite au hasard, donc un coefficient directeur *a* au hasard et un OO *b* au hasard
- repeter:
    - prendre un point du dataset
    - mettre a jour la droite pour se rapprocher du point en utilisant la metode square ou absolue
    
Chaque tour du bloc "repeter" s'appelle une "epoch". Au debut de notre entrainement il faut specifier pendant combien d'epoch on souhaite entrainer le modele pour savoir quand s'arreter. Puisque la metode "square" est plus populaire on va partir sur celle la pour la suite.

Voici le code

```python
import random
def linear_regression(features, labels, learning_rate=0.01, epochs = 1000):
    price_per_room = random.random()
    base_price = random.random()
    for epoch in range(epochs):
        i = random.randint(0, len(features)-1)
        num_rooms = features[i]
        price = labels[i]
        price_per_room, base_price = square_technique(base_price,
                                                  price_per_room,
                                                  num_rooms,
                                                  price,
                                                  learning_rate=learning_rate)
    return price_per_room, base_price
```

Il nous faut encore importer proprement le dataset

```python
import numpy as np
features = np.array([1,2,3,5,6,7])
labels = np.array([155, 197, 244, 356, 407, 448])
```

```python
linear_regression(features, labels, learning_rate = 0.01, epochs = 10000)
```
