## Functions

On a vu que les variables nous permettent d'associer une valeur √† un nom. Le but d'une fonction va √™tre de pouvoir associer une suite d'instructions √† un nom, pour les m√™mes raisons que les variables, lisibilit√© et reutilisabilit√©.

````python
# on d√©marre par le mot clef "def" pour signaler √† l'ordinateur que ce qui suit va √™tre une nouvelle fonction
# ensuite on lui donne un nom de notre choix comme "sum_of_two_plus_some_stuff"
# ensuite entre les paranth√®ses nous avons les "param√®tres".
# les param√®tres sont des variables temporaires qui ne durent que le temps de l'ex√©cution de la fonction et qui contiennent les entr√©es de la fonction
# en maths on utilisait la notation "f(x) => 3x+2" par example, ici "x" est un param√®tre et dans le cas f(3) le param√®tre x vaux 3 alors que dans le cas f(12) x vaux 12
# les deux points ":" marquent le d√©but des instructions contenues dans la fonction
def sum_of_two_plus_some_stuff(first, second):
    result = first + second# on se cr√©er une variable result qu'on va transformer √† mesure qu'on avance
    result = result + 3
    result = result + 5
    result = result - 7
    return result# le mot clef return nous laisse designer le resultat de la fonction
                 # il est tout √† fait possible de ne pas avoir de r√©sultat, le return est optionnel
#notez que nous somme revenu un cran vers la gauche, marquant la fin de la function

def do_nothing(): # une fonction qui ne prend pas de param√®tre √† quand meme des paranth√®ses
    pass
#fin

# cette ligne va donc ex√©cuter le code contenu dans la fonction "sum_of_two_plus_some_stuff" 
# first vaudra 12 et second vaudra 25
sum_of_two_plus_some_stuff(12,25)# le r√©sultat de cette ligne √©tant donc 38

# cette ligne va donc ex√©cuter le code contenu dans la fonction "sum_of_two_plus_some_stuff" 
# first vaudra 2 et second vaudra 5
sum_of_two_plus_some_stuff(2,5)# le r√©sultat de cette ligne etant donc 8

# cette ligne va donc ex√©cuter le code contenu dans la fonction "sum_of_two_plus_some_stuff" 
# first vaudra 2 et second vaudra 5
sum_of_two_plus_some_stuff(-2,-5)# le r√©sultat de cette ligne etant donc -6

# on peut stocker le r√©sultat d'une fonction dans une variable pour s'en servir plus tard
some_number_I_dont_care_about = sum_of_two_plus_some_stuff(-2,-5)
````

## Contr√¥le de l'ex√©cution
### if, else et elif

Vous pouvez choisir d'avoir du code qui s'ex√©cute ou non en fonction de certaines conditions grace aux blocs "if". Si vous voulez couvrir plus de cas vous avez √©galement les blocs "else" et "elif" qui sont compl√©mentaires. Les bloc "else" ne peuvent √™tre ecrits qu'en dernier et "elif" sont √† mettre apr√®s le "if"

````python
x = 16

if x < 0: # d'abord on √©crit if puis la condition √† remplir et enfin les ":" pour d√©limiter le d√©but du bloc
    x = x / 1024
    x = x + 3

# notez qu'on reviens une indentation vers la gauche pour fermer le bloc if

if x > 0:
    x = 15
else:      # si jamais vous voulez faire quelque chose quand la condition du if N'EST PAS valide,
    x = 0  # vous pouvez ecrire un bloc "else" √† la fin du "if", on ex√©cutera le code du bloc else seulement si celui du if n'as pas √©t√© ex√©cut√©
    
if x == 1: # vu qu'un seul √©gal "=" est d√©j√† utilis√© pour changer la valeur d'une variable il faut utiliser un double √©gal "==" pour checker l'√©galit√©
    x = "hun"
elif x == 2:   # elif check une nouvelle condition si le if ci-dessus n'est pas valide
    x = "zwei"  
elif x == 3:                         # ce elif check encore une nouvelle condition si aucune des deux conditions n'ont √©t√© valid√©es
    x = "The Great Jecht Shoot No.3" # notez que pour la premi√®re condition valid√©e on ex√©cute le bloc et on passe √† le suite de TOUTE la cha√Æne de if/elif/else
else:                         
    x = "we're hitting rock bottom, and still digging cap'"
#fin de la totale, notez que les if, elif et else sont align√©s pour marquer le debut de leurs blocs respectifs
````

### for
L'instruction "for" permet d'ex√©cuter plusieurs fois un seul bloc pour chaque √©l√©ment d'une liste. (...plus exactement d'un "it√©rable" plut√¥t qu'une liste mais pour le moment on ne fera pas la distinction)

Il √©xiste un fonction qui permet de g√©n√©rer un it√©rable de nombres croissant appell√©e "range()". On peut se servir de for et range() pour √©xecuter un bloc de code un certain nombre de fois. Ex√©cutez le code ci-dessous pour voir ce que √ßa donne.

````python
for i in range(13): # "pour chaque √©l√©ment que l'on appellera i dans la range allant de 0 √† 13"...
    print(i)        # ... afficher i
````

Quelque chose de particuli√®rment pratique est que √ßa nous permet d'√©x√©cuter du code sur tous les √©l√©ments d'une liste

````python
some_list = [1, 4.2, 17, -12, 72, 2079, -20.11]
total = 0

for e in some_list:
    total = total + e
    
print(total)
````
