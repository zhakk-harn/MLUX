#### L'apprentissage non-supervisé

L'apprentissage non-supervisé se fait sur des données non "labélisées". Le but de l'algorithme est de trouver de l'information pertinente sans pouvoir s'appuyer sur un jeu de "bonnes réponses". ##### Clustering n'est pas classification

Notez que le clustering rassembles des éléments similaires, tandis que la classification rassemble des éléments sous le même label que vous avez vous même définis. C'est pour ça que dans l'exemple ci-dessus les emails n'ont pas de signification particulière, ils sont similaires mais un email du cluster 3 ne garantis pas qu'il s'agisse d'un email promotionnel, d'un spam, d'une lettre de départ ou d'un memo, juste qu’il ressemble aux autres mails du cluster 3 au regards des infos contenues dans la dataset. Pour trier par types il faudrait utiliser un algo de classification entrainé avec des exemples pour chaque type d'email.

## Régressions linéaire

La régression linéaire est une méthode efficace et très répandue pour estimer des valeurs comme la valeur d'une maison, la valeur d'une action en bourse, l’espérance de vie d'un individu, ou le temps qu'un utilisateur va passer à regarder une vidéo sur votre site web. On peut voir la régression linéaire sous une forme plus graphique que mathématique. C'est ce que l'on va faire en visualisant des points et une ligne qui bouge.

Ces points forment vaguement une ligne.
<img src="../assets/img/regression_lineaire/dots.png"/>

Le but de la régression linéaire est de tracer une droite qui passe au plus proche de tous les points. Un peu comme ça :

On peut commencer à se poser des questions :
- En quoi c'est utile ?
- En quoi est-ce que c'est du machine Learning ?
Pour ça on va voir comment utiliser la régression linéaire pour prédire le prix de maisons via un dataset.

Si on veut vendre une maison, le mieux serais de se baser sur les autres maisons qui sont sur le marché. On peut regarder leurs features qui pourrais influencer le prix de la maison comme sa taille, le nombre de pièces, son emplacement, les écoles et les commerces à proximité. Au final on voudrait juste une formule qui nous rend le prix que vaux notre maison ou au moins une estimation plutôt correcte.

On va faire ça avec un exemple très simple. On va seulement se baser sur le nombre de pièces. Notre maison en a 4 et il y a 6 maisons à vendre dans le coin qui ont une, deux, trois, cinq, six et sept pièces.

| pièces | prix |
|--------|------|
|  1      |   150   |
|  2      |   200   |
|  3      |   250   |
|  4      |   ??   |
|  5      |   350   |
|  6      |   400   |
|  7      |   450   |

D'instinct vous allez dire que la maison avec 4 pièces vaut 300. Et vous auriez raison. On peut voir une logique et faire une déduction basée dessus. Ce que vous venez de faire de manier intuitive est une régression linéaire. On peut remarquer qu'à chaque fois qu'on rajoute une pièce on rajoute 50 au prix, en fait on peut voir le prix de base de la maison comme étant de 100 puis de 50 par pièce, ce qu'on pourrait écrire comme ça 

Prix = 100 + 50*(nombre de pièces)

Ce qu'on vient de faire avec cette écriture est de spécifier un *modèle* qui donne une *prédiction* basée sur une *feature*, ici le nombre de pièces. Si on refait une tour rapide des définitions ça nous donne
> **feature**  
Les features d'un point de donnée sont les propriétés sur lesquelles on va s’appuyer pour faire nos prédictions. Ici on en a qu'une seule, le nombre de pièces dans la maison

> **label**  
La feature que l'on essaye de prédire. Dans ce cas c'est le prix de la maison.

> **modèle  / model**  
La règle, ou la formule, qui prédit un label grâce aux features. Dans ce cas, le modèle est la formule qui nous permet de calculer le prix de la maison.

> **prédiction**  
Le résultat donné par le modèle, ici le modèle donne 300 comme prix si le nombre de pièces est de 4 donc "300" est la prédiction.

Maintenant la question est "Comment est-ce qu'on peut faire pour que l'ordinateur trouve cette formule ?". Pour mieux illustrer ça on va prendre un exemple un peu moins direct. On va aussi déconstruire le processus en utilisant les étapes "remember-formulate-predict" vu plus tôt. On va *remember* le prix des maisons, *formulate* un modèle puis utiliser le modèle pour *predict* le prix de n'importe quelle maison.

### Formuler une règle qui estime le prix des maisons
La formule utilisée pour le dataset précèdent devrait encore être correct quoiqu’un peu faux donc on va l'écrire comme ça

Prix = 100 + 50*(nombre de pièces) + (marge d'erreur)

On peut voir qu'à chaque fois qu'on se déplace d'un x vers la droite la ligne formée par les maisons monte de 50 et que cette ligne traverse l'axe y a la valeur 100. On dit que 50 est le *coefficient directeur* et que 100 est l'*ordonnée à l'origine*. Pour faire plus court j'utiliserais "pente" et "OO"

> Coefficient directeur / pente  
Le coefficient directeur dicte l'angle que prend la ligne. On peut le retrouver en regardant de combien la ligne monte ou descend lorsque l'on se déplace d'une unité vers la droite. Si la ligne descend le coefficient sera négatif, si elle monte il sera positif. Si la ligne est horizontale le coefficient est de 0. Ça nous dira à quel point la valeur du label va changer si on change la valeur de la feature.

> Ordonnée a l'origine / OO  
L'ordonnée a l'origine est la valeur à laquelle la droite intersecte avec l'axe vertical (les ordonnées). Il s'agit de la valeur du label lorsque la feature est absente.

> Equation linéaire  
Il s'agit de l'équation représentant une ligne, elle a 2 composantes, un coefficient directeur et une ordonnée à l'origine. Pour une pente *a* et une ordonnée à l'origine *b* on peut représenter la ligne comme ceci y = ax + b. En machine learning, *x* est la valeur de la feature et *y* est la prédiction pour le label.

### Qu'est-ce qu'il se passe si on veut utiliser plus qu'une feature ? La régression multilinéaire
Si on imagine qu'on veut prédire le prix d'une maison en se basant pas seulement sur le nombre de pièce mais aussi sa taille, la proximité avec des écoles et l’âge de la maison notre modèle de régression linéaire reste utile. Pour ça il faut utiliser chaque feature avec sa propre pente ce qui nous donnerais un modèle qui pourrais ressembler à ça  

Prix = 30(nombre de pièces) + 1.5(taille) + 10(proximité des écoles) - 2(âge) + 50

Tout est positif sauf le coefficient devant l’âge de la maison ? Parce que plus l’âge de la maison est important plus son prix baisse, le coefficient est donc négatif. On dit que la corrélation entre le prix et l’âge est négative tandis que pour les autres paramètres elle est positive.  
Qu'est-ce qu'il se passe si le coefficient est de 0 plutôt que 10, 1.5 ou 30 ? Ça voudrait dire que la feature en question n'as pas d'influence sur le prix. Au contraire plus le coefficient est un grand nombre et plus la feature est influente.

Pendant ce chapitre on va visualiser notre modèle comme une droite mais pendant la visualisation chaque colonne de notre dataset représente une dimension donc si on a un dataset avec deux colonnes on peut le représenter sur un graph avec 2 axes. Un dataset avec 3 colonnes sera représenté comme une collection de points dans un graph a 3 axes, donc en 3 dimensions, à ce stade le modèle sera représenté comme un plan qui doit se rapprocher au maximum de tous les points. Un peu comme si j'avais quelques mouches en vol stationnaire et que je devais placer une planche de carton aux plus proche de chacun d’entre eux. Le problème est quand on essaye de se représenter la régression multilinéaire sur un dataset avec 4 colonnes ou plus. C'est toujours calculable mais plus visualisable.

Pour le moment on va s'en tenir à un dataset avec une seul feature mais en principe la méthode est la même si vous en aviez plusieurs.

1. Qu'est-ce qu'il se passe si le modèle se trompe ?
Le but du model est de fournir une estimation du prix, on s'attend à ce qu'il ne soit pas correct mais le processus d'entrainement devrait rendre ces écarts suffisamment petit.
1. Comment on ferait pour trouver un modèle si on avait des milliers de maisons et pas seulement 6
A vu de nez c'est difficile de tracer une droite qui passe au plus proche de tous les points c'est pour ça qu'on va voir comment utiliser une machine pour faire ce travail.
1. Si on génère le modèle mais que plus tard de nouvelles maisons arrivent sur le marché, est ce qu'on peut mettre à jour le modèle avec ces nouvelles informations ?
Oui, pour peux que la relation entre les features et le label continues d'être valides on aura pas besoin de retoucher à l'algorithme, il faudra juste mettre à jour le dataset et régénérer le modèle.

Pour ce qui est de retrouver la ligne/modèle via l'ordinateur on va Il n’y prendre pas à pas, comme très souvent en machine learning. On va commencer par utiliser une ligne complètement aléatoire (ou avec une pente et une OO nulles) et on va la rapprocher des points petits à petit jusqu'à avoir une ligne satisfaisante.

La version simpliste est de prendre un point au hasard et de rapprocher la ligne, puis refaire la même avec un autre point et ainsi de suite. Ce qui nous donnerais la procédure suivante :  

Entrées : un dataset de points
Sortie : une ligne qui passe proche des points
Procédure :
- prendre une ligne au hasard
- répéter :
    - prendre un point du dataset au hasard 
    - bouger la ligne pour la rapprocher du point
- résultat : la ligne

Si on répète suffisamment de fait le rapprochement de la ligne on se retrouvera avec une ligne plutôt proche de tous les points.

<img src="../assets/img/regression_lineaire/point_by_point.png"/>

Pour rentrer un peu plus dans le détail on va recommencer à utiliser des formules. Pour faire plus simple on va définir les raccourcis suivant
- y: le prix d'une maison dans le dataset
- y': la prédiction du prix pour une maison
- x : le nombre de pièces
- a : le prix d'une pièce
- b : le prix de base d'une maison avant de compter les pièces

On retrouve bien l'équation de notre droite

y' = ax + b

Ce qui revient toujours à 

Prédiction du prix = (prix par pièce) (nombre de pièces) + prix de bases avant les pièces

Pour commencer à se faire une idée de la procédure de la régression linéaire, prenons un modèle dans lequel les pièces valent 40 et le prix de base d'une maison est de 50 nous donnants y' = 40x + 50  
Imaginons maintenant que notre dataset contienne une maison avec 2 pièces qui coute 150. Ce modèle prédire qu'elle vaut 40\*2+50=130. Ce n’est pas loin mais c'est pas ça. Vu que l'estimation tombe en dessous on pourrait améliorer le modèle en augmentant un peu le prix des pièces et le prix de base d'une maison. Si on augmente le prix par pièce de 0.50 et le prix de base de 1 ça nous donne un nouveau modèle qui prédit 40.5\*2+51=132.  
C'est plus proche de 150, l'erreur est plus faible et donc le modèle est meilleur ... pour ce point de donnée seulement. 

Si on veut pouvoir rapprocher une ligne il va falloir comprendre comment la manipuler. On disait que le coefficient représente la montée ou la descente de la ligne quand augment sur l'axe des x (on va à droite). Si mon coefficient directeur est de 0.5, je m’attends à ce que pour chaque x ma ligne monte de 0.5y et si mon coefficient directeur est de 2 je m’attends à ce que pour chaque x ma ligne monte de 2y. Autrement dit, augmenter le coefficient directeur fait tourner la droite dans le sens contraire des aiguille d'une montre et a l'inverse le réduire fait tourner la droite dans le sens des aiguilles d'une montre.

Si l'ordonnée a l'origine est l'intersection entre la ligne et l'axe y alors l'augmenter revient à décaler la droite vers le haut et le réduire reviens à décaler la ligne vers le bas.

Autrement dit, l'OO nous donne la position de la ligne et le coefficient directeur nous donne sa direction.

### La version naïve
On disait que notre algorithme de régression linéaire consistait à continuellement rapprocher la ligne d'un point puis d'un autre et ainsi de suite. On va donc voir comment faire tourner et glisser la droite plus près dans tous les différents cas.

<img src="../assets/img/regression_lineaire/closer.png"/>

Pour ça il va d'abord falloir trouver ou est le point par rapport à la droite. Si le point est sous la ligne il va falloir glisser vers le bas, s’il est au-dessus il va falloir glisser vers le haut. Pour tourner ça va être un peu plus complique parce que la ligne tourne sur son OO. On va avoir besoin de savoir si le point est au-dessus à gauche ou en dessous à droite de l'OO, ce qui nous indiquerais qu'il faut tourner dans le sens des aiguilles d'une montre. A l'inverse si le point est au-dessus à droite ou en dessous à gauche il va falloir tourner dans le sens contraire. Donc :  

- Cas 1 : Le point est au-dessus de la ligne et à gauche de l'axe y, on tourne dans le sens antihoraire et on glisse vers le haut
- Cas 2 : Si le point est au-dessus de la ligne et à droite de l'axe y, on tourne dans le sens horaire et on glisse vers le haut
- Cas 3 : Si le point est en dessous de la ligne et à droite de l'axe y, on tourne dans le sens horaire et on glisse vers le bas
- Cas 4 : Si le point est en dessous de la ligne et à gauche de l'axe y, on tourne dans le sens antihoraire et on glisse vers le bas

<img src="../assets/img/regression_lineaire/rotate_translate.png"/>

Maintenant que ça c'est clair, on va pouvoir détailler notre algorithme

Pour rappel notre droite ressemble à : (y' = ax+b) où :
- y' est la prédiction du prix d'une maison
- a est le coefficient directeur de la droite et le prix d'une pièce
- x est le nombre de pièce pour lequel on souhaite avoir une prédiction
- b est l'ordonnée à l'origine soit le prix de la maison avant de prendre en compte le nombre de pièces
- (x,y) est un point du dataset qui fait correspondre un prix y a un nombre x de pièces

Entrées : 
- une ligne ayant une pente a, et une OO b, d'équation y' = ax+b
- un point de coordonne (x,y)
Sortie : 
- une ligne d'équation y' = a'x+b' ou a' et b' sont un nouveau coefficient directeur et OO qui rapproche la droite du point
Procédure :
- prendre deux petit nombres n1 et n2
- si : le point est au-dessus de la ligne et à droite de l'axe y
    - ajouter n1 a la pente tel que a' = a + n1
    - ajoute n2 a l'OO tel que b' = b + n2
- si : le point est au-dessus de la ligne et à gauche de l'axe y
    - retirer n1 a la pente tel que a' = a - n1
    - ajouter n2 a l'OO tel que b' = b + n2
- si : le point est en dessous de la ligne et à droite de l'axe y
    - retirer n1 a la pente tel que a' = a - n1
    - retirer n2 a l'OO tel que b' = b - n2
- si : le point est en dessous de la ligne et à gauche de l'axe y
    - ajouter n1 a la pente tel que a' = a + n1
    - retirer n2 a l'OO tel que b' = b - n2
- résultat : y' = a'x + b'

Dans notre exemple modifier *a* reviens à modifier le prix d'une pièce te modifier *b* reviens à modifier le prix de base d'une maison. Vu que le nombre de pièces ne peux pas être négatif on peut simplifier la procédure en oubliant les cas 1 et 3.

En principe cette méthode fonctionne et elle a le mérite d'être simple mais ce n'est pas ce qui est utilisé car elle n'est vraiment pas efficace. En plus de ça :
- comment est-ce que l'on choisit des bonnes valeurs pour n1 et n2 ?
- est ce qu’on ne pourrait pas remplacer les 4 cas par une même étape qui marche tous le temps ?
On peut ! Et pour ça on va voir les méthodes qu'on appellera "square" pour le moment.

### la version "square"
Au final le but de nos 4 cas était de trouver le bon signe pour (+ ou -) pour changer les paramètres *a* et *b* de la droite pour se rapprocher du point.

Commençons avec le déplacement. Si le point (x,y) est au-dessus de la droite, alors y - y' sera positif parce que y > y', en revanche si le point est en dessous alors y - y' sera négatif parce que y < y'.

<img src="../assets/img/regression_lineaire/prediction_delta.png"/>

Donc on peut juste ajouter (y - y') a l'OO *b* pour faire glisser la droite dans la bonne direction. Le problème est que (y - y') est possiblement trop grand et résultera en un glissement trop violent et donc à ne jamais stabiliser une droite correcte. C'est pour ça qu'on va devoir parler de "rythme d'apprentissage"

> Rythme d'apprentissage / Learning Rate  
Un petit nombre qu'on sélectionné avant d'entrainer notre modèle. Ce nombre nous permet de réduire la taille des changements à chaque étape. On le notera *n*

Parce que *n* est petit on peut maintenant ajouter n(y-y') a l'OO *b* sans problème.

Pour ce qui est de modifier la pente c'est très similaire.  
Un point (x,y) est à gauche de l'axe y si *x* < 0 et à droite si *x* > 0

<img src="../assets/img/regression_lineaire/x_to_y.png"/>

Donc x(y-y') est positif si *x* et (y-y') sont tous les deux positifs ou tous les deux négatifs comme *a'* dans les cas 1 et 4. Tandis que si seulement *x* ou seulement (y-y') est positif alors x(y-y') sera négatif comme *a'* dans le cas 2 et 3. Du coup si on ajoute x(y-y') aux coefficient directeur *a* alors la droite tournera dans le bon sens. On n’oublie pas de le multiplier par le learning rate pour avancer a pas mesures et on est bon. nx(y-y') nous fera toujours tourner la droite dans le bon sens.

Ce qui nous donne la nouvelle procédure 
Entrées : 
- une ligne ayant une pente a, et une OO b, d'équation y' = ax+b
- un point de coordonné (x,y)
- un petit nombre positif *n* (le learning rate)
Sortie : 
- une ligne d'équation y' = a'x+b' ou a' et b' sont un nouveau coefficient directeur et OO qui rapproche la droite du point
Procédure :
- ajouter nx(y-y') au coefficient directeur pour mettre à jour la direction de la droite, a' = a + nx(y-y')
- ajouter n(y-y') a l'OO pour déplacer la droite, b' = b + n(y-y')
- résultat : y' = a'x + b'

Maintenant on va pouvoir coder ça pour l'utiliser


```python
def square_technique(base_price, price_per_room, num_rooms, price, learning_rate):
    predicted_price = base_price + price_per_room*num_rooms
    base_price += learning_rate*(price-predicted_price)
    price_per_room += learning_rate*num_rooms*(price-predicted_price)
    return price_per_room, base_price
```

### alternative : la version absolue
Il existe une autre méthode qu'on appellera "absolue" qui est tout aussi utilisable et un peu plus simple que la square.  
Dans la square on utilise y-y' (le différence entre la prédiction et la vraie valeur) et x (le nombre de pièces) pour couvrir tous les cas en un. Mais on peut se baser uniquement sur x pour passer de 4 cas a 2. Voici la procédure

Entrées : 
- une ligne ayant une pente a, et une OO b, d'équation y' = ax+b
- un point de coordonné (x,y)
- un petit nombre positif *n* (le learning rate)
Sortie : 
- une ligne d'équation y' = a'x+b' ou a' et b' sont un nouveau coefficient directeur et OO qui rapproche la droite du point
Procédure :
- si : le point est au-dessus de la ligne
    - ajouter nx au coefficient directeur, a' = a + nx (on tourne dans le sens antihoraire si le point est à droite de l'axe y ou dans le sens horaire si x est à gauche de l'axe y)
    - ajouter n à l'OO tel que b' = b + n
- si : le point est en dessous de la ligne
    - retirer nx à la pente tel que a' = a - nx (on tourne dans le sens horaire si le point est à droite de l'axe y ou dans le sens antihoraire si est à droite de l'axe y) 
    - retirer n à l'OO tel que b' = b - n
- résultat : y' = a'x + b'

Et voici le code qui va avec

```python
def absolute_technique(base_price, price_per_room, num_rooms, price, learning_rate):
   predicted_price = base_price + price_per_room*num_rooms
   if price > predicted_price:
       price_per_room += learning_rate*num_rooms
       base_price += learning_rate
   else:
       price_per_room -= learning_rate*num_rooms
       base_price -= learning_rate
   return price_per_room, base_price
```

### L'algorithme de régression linéaire : répéter le square ou absolute
Le plus dur est fait. Il nous reste plus qu'à prendre une droite au hasard et tous les points pour les faire passer en boucle dans l'une des deux techniques qu'on vient de voir jusqu'à ce que la ligne soit bien ajustée.  

Voilà la procédure

Entrées : 
- un dataset de maisons ayant chacune un prix et un nombre de pièces
Sortie : 
- un modèle pouvant prédire le prix en fonction du nombre de pièces
Procédure :
- prendre une droite au hasard, donc un coefficient directeur *a* au hasard et un OO *b* au hasard
- répéter :
    - prendre un point du dataset
    - mettre à jour la droite pour se rapprocher du point en utilisant la méthode square ou absolue
    
Chaque tour du bloc "répéter" s'appelle une "epoch". Au début de notre entrainement il faut spécifier pendant combien d'epoch on souhaite entrainer le modèle pour savoir quand s'arrêter. Puisque la méthode "square" est plus populaire on va partir sur celle-là pour la suite.

Voici le code

```python
import random
def linear_regression(features, labels, learning_rate=0.01, epochs = 1000):
    price_per_room = random.random()
    base_price = random.random()
    for epoch in range(epochs):
        i = random.randint(0, len(features)-1)
        num_rooms = features[i]
        price = labels[i]
        price_per_room, base_price = square_technique(base_price,
                                                  price_per_room,
                                                  num_rooms,
                                                  price,
                                                  learning_rate=learning_rate)
    return price_per_room, base_price
```

```python
import numpy as np
features = np.array([1,2,3,5,6,7])
labels = np.array([155, 197, 244, 356, 407, 448])
```

```python
linear_regression(features, labels, learning_rate = 0.01, epochs = 10000)
```
